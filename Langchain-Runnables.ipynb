```python
import config_api
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
chat = ChatOpenAI(
    api_key = config_api.api_key,
    model = 'gpt-4.1-mini',
    max_tokens = 100
)
```

```python
str_output_parser = StrOutputParser()
```

```python
chat_prompt_1 = ChatPromptTemplate.from_template('''
What are the five most important tools a {job title} needs?
Answer only by listing the tools.
''')

chat_prompt_2 = ChatPromptTemplate.from_template('''
Considering the tools provided, develop a strategy for effectively learning and mastering them:
{tools}
''')
```

```python
chain_1 = chat_prompt_1 | chat | str_output_parser | {'tools': RunnablePassthrough()}
chain = chain_1 | chat_prompt_2 | chat | str_output_parser
print(chain.invoke({'job title:' 'AI Engineer'}))
```

```python
chain.get_graph().print_ascii()
```

```python
from langchain_core.runnables import RunnableParallel
```

```python
chat_prompt_template_books = ChatPromptTemplate.from_template('''
Suggest three of the best intermediate-level {programming language} books. 
    Answer only by listing the books.
''')

chat_template_projects = ChatPromptTemplate.from_template(
    '''
    Suggest three interesting {programming language} projects suitable for intermediate-level programmers. 
    Answer only by listing the projects.
    '''
)
```

```python
chain_books = chat_prompt_template_books | chat | str_output_parser
chain_projects = chat_template_projects | chat | str_output_parser

chain_parallel = RunnableParallel({'books': chain_books, 'projects': chain_projects })

chain_parallel.invoke({ 'programming language': 'Python'})
```

```python
chain_parallel.get_graph().print_ascii()
```

```python
chat_template_time = ChatPromptTemplate.from_template('''
I'm an intermediate level programmer.
     
     Consider the following literature:
     {books}
     
     Also, consider the following projects:
     {projects}
     
     Roughly how much time would it take me to complete the literature and the projects?
''')
```

```python
chain_time = ({'books': chain_books,'projects': chain_projects}) | chat_template_time | chat | str_output_parser
```

```python
chain_time.invoke({'programming language': 'Python'})
```

```python
from langchain_core.runnables import RunnableLambda
```

```python
find_sum = lambda x: sum(x)
find_sum([1,2,3])
```

```python
find_square = lambda x: x ** 2
find_square(6)
```

```python
runnable_sum = RunnableLambda(lambda x: sum(x))

runnable_square = RunnableLambda(lambda x: x ** 2)

chain = runnable_sum | runnable_square

chain.invoke([1,2,3])
```

```python
from langchain_core.runnables import chain
```

```python
def find_sum1(x):
    return sum(x)

def find_square1(x):
    return x**2
```

```python
chain1 = RunnableLambda(find_sum1) | RunnableLambda(find_square1)
chain1.invoke([1,2,3])
```

```python
@chain
def runnable_sum1(x):
    return sum(x)

@chain
def runnable_square1(x):
    return x**2
```

```python
chain2 = runnable_sum1 | runnable_square1
chain2.invoke([1,2,3])
```
