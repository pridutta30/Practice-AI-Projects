```python
import config_api
```

```python
api_key = config_api.api_key
```

```python
from openai import OpenAI
```

```python
client = OpenAI(api_key=api_key)
```

```python
def sarcastic_chatbot(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input = [
            {
                'role': 'system',
                'content': 'Respond sarcastically to the questions asked'
            },
            {
                'role': 'user',
                'content': 'What can I do for the weekend?'
            },
            {
                'role': 'user',
                'content': prompt
            }
        ],
        stream = True,
        max_output_tokens = 50
    )
    full_response = ""
    for event in response:
        if event.type == "response.output_text.delta":
            print(event.delta, end="", flush=True)
            full_response += event.delta

    return full_response
```

```python
prompt = 'How are you feeling today'
sarcastic_chatbot(prompt)
```

```python
def tweet_classifier(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input = [
            {
                "role": 'system',
                "content": "Act as a sentiment analyzer and respond to the tweets as positive, negative or neutral"
            },
            {
                "role": 'user',
                "content": "I hate mangoes"
            },
            {
                "role": 'assistant',
                "content": "Negative"
            },
            {
                "role": 'user',
                "content": "What a beautiful day"
            },
            {
                "role": 'assistant',
                "content": "Positive"
            },
            {
                "role": 'user',
                "content": "Tomorrow is Friday"
            },
            {
                "role": 'assistant',
                "content": "Neutral"
            },
             {
                "role": "user",
                "content": prompt
            }
        ]
    )
    return response.output_text.strip()
    
```

```python
prompt = "The movie was terrible"
tweet_classifier(prompt)
```

```python
prompt = "World paying a heavy price for US Iran war"
tweet_classifier(prompt)
```

```python
prompt = "weekends are the best!"
tweet_classifier(prompt)
```

```python
prompt = "is it raining?"
tweet_classifier(prompt)
```
