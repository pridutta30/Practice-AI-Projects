```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import CommaSeparatedListOutputParser
import config_api
```

```python
api_key = config_api.api_key
```

```python
chat = ChatOpenAI(api_key = api_key,
                 model = 'gpt-4.1-mini',
                 max_tokens = 150,
                 seed = 365,
                 temperature = 0)
```

```python
list_output_instructions = CommaSeparatedListOutputParser().get_format_instructions()
```

```python
chat_template = ChatPromptTemplate.from_messages(['human',
                                                 'Suggest four {pet} names \n' + list_output_parser])
```

```python
chat_value = chat_template.invoke({'pet':'dog'})
```

```python
response = chat.invoke(chat_value)
```

```python
list_output_parser = CommaSeparatedListOutputParser()
list_output_parser.invoke(response)
```

```python
chain = chat_template | chat | list_output_parser
chain.invoke({'pet':'cat'})
```

```python
chain.batch([{'pet':'dog'},{'pet':'cat'},{'pet':'bird'},{'pet':'fish'}])
```
